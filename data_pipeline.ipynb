## Popularily Based Recommendation System

```python
df = ratings.merge(books, on = 'isbn')
df.head()
```

```python
df.shape
```

```python
no_rating_df = df.groupby('book_title').count()['book_rating'].reset_index()
no_rating_df.rename(columns = {'book_rating': 'no_of_rating'}, inplace = True)
no_rating_df
```

```python
avg_rating_df = df.groupby('book_title', as_index=False)['book_rating'].mean()
avg_rating_df.rename(columns = {'book_rating': 'avg_of_rating'}, inplace = True)
avg_rating_df['avg_of_rating'] = avg_rating_df['avg_of_rating'].round(2)
```

```python
avg_rating_df
```

```python
# clamp the rating at a maximum of 5.0
avg_rating_df['avg_of_rating'] = avg_rating_df['avg_of_rating'].clip(upper=5.0)
avg_rating_df
```

```python
popular_df = no_rating_df.merge(avg_rating_df, on = 'book_title')
popular_df
```

```python
popular_df = popular_df[popular_df['no_of_rating']>=250].sort_values('avg_of_rating', ascending = False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books, on = 'book_title').drop_duplicates('book_title')[['book_title', 'book_author', 'image_url_l', 'no_of_rating', 'avg_of_rating']]
```

```python
popular_df
```

## Collaborative Filtering Based Recommender System
Returning rows where the users have rated more than 200 books and books with minimum of 50 ratings

```python
x = df.groupby('user_id').count()['book_rating']>200
users_rated_books = x[x].index
```

```python
filtered_rating = df[df['user_id'].isin(users_rated_books)]
```

```python
y = filtered_rating.groupby('book_title').count()['book_rating']>=50
popular_books = y[y].index
```

```python
final_df = filtered_rating[filtered_rating['book_title'].isin(popular_books)]
```

```python
pt_df = final_df.pivot_table(index = 'book_title', columns = 'user_id', values = 'book_rating')
```

```python
pt_df.fillna(0, inplace = True)
```

```python
pt_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt_df)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    # fetch index
    index = np.where(pt_df.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x:x[1], reverse = True)[1:11]

    data = []
    for i in similar_items:
        temp_df = books[books['book_title'] == pt_df.index[i[0]]]
        item = []
        item.extend(list(temp_df.drop_duplicates('book_title')['book_title'].values))
        item.extend(list(temp_df.drop_duplicates('book_title')['book_author'].values))
        item.extend(list(temp_df.drop_duplicates('book_title')['image_url_l'].values))

        data.append(item)

    return data
```

```python
recommend('Harry Potter and the Prisoner of Azkaban (Book 3)')
```
